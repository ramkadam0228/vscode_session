```python
import numpy as np
np.random.seed(42)
x1=np.random.rand(1000)
y1=2*x1+np.random.rand(1000)*0.7
x2=np.random.rand(1000)
y2=-2*x2+np.random.rand(1000)*0.7
plt.scatter(x1,y1, color='blue',alpha=0.3,label="Positive correlation")
plt.scatter(x2,y2, color='red',alpha=0.3,label="Negative correlation")
plt.legend()
plt.show()
```
