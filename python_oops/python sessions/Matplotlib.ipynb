```python
import pandas as pd # Dataframe
import numpy as np  # Math
import matplotlib.pyplot as plt # Plotting
import seaborn as sns # Plotting
import random
visa_df=pd.read_csv('Visadataset.csv')
cat_col=visa_df.select_dtypes(include=['object'])
cat_cols=cat_col.columns
cat_cols
```

```python
# Saving plot in folders
# lets create a folder first
# i want to create a folder one step ahead from my current notebook
# we can create folder using python itself
# for that we need to provide the folder name
# using the os package
 
import os
new_folder_name='plots'
try:
    os.makedirs(new_folder_name)
except Exception as e:
    print(e)
cwd=os.getcwd()
new_path=os.path.join(cwd,new_folder_name)
 
for i in range(len(cat_cols)):
    info=visa_df[cat_cols[i]].value_counts()
    index=info.index
    values=info.values
    colors=random.sample(colors_list,len(index))
    plt.bar(index,values,color=colors)     
    plt.xlabel(cat_cols[i])   
    plt.ylabel('count') 
    plt.title('bar chart') 
    plt.savefig(f"{new_path}\{cat_cols[i]}.jpg")
    plt.show()
```

```python
import pandas as pd
import pandas as pd # Dataframe
import numpy as np  # Math
import matplotlib.pyplot as plt # Plotting
import seaborn as sns # Plotting
import random

visa_df=pd.read_csv('Visadataset.csv')
visa_df
visa_df['continent'].value_counts()
keys=visa_df['continent'].value_counts().keys()
values=visa_df['continent'].value_counts().values
print(keys, values)
plt.pie(x=values, labels=values)
plt.pie(x=values, labels=keys, autopct="%0.2f%%")
plt.pie(x=values, labels=values, autopct="%0.2f%%")
plt.show()
```
