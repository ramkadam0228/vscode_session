```python
# Video

import cv2
vid=cv2.VideoCapture(0)
while True:
    ret,frame=vid.read()
    cv2.imshow("Window_name: Frame", frame)
    if cv2.waitKey(1) & 0xFF==ord("q"):
        break

vid.release()
cv2.destroyAllWindows()
```

```python
#BGR = 0,0, 255 Since this is computer vision
# BGR=0,255,0
# BGR=255,0,0


import cv2
vid=cv2.VideoCapture(0)
while True:
    ret,frame=vid.read()
    font=cv2.FONT_ITALIC
    cv2.putText(frame,
                "hellow Ram"
                ,(50,50)
                ,font
                ,2
                ,(0,0,255))
    
    cv2.imshow("Window_name: Frame", frame)
    if cv2.waitKey(1) & 0xFF==ord("q"):
        break

vid.release()
cv2.destroyAllWindows()
```
