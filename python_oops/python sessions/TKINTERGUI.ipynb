```python
# Step 5 ############### Adding check button ####################

# Step # 3
# Add Button
import tkinter as tk
# create the main window and add label
root=tk.Tk()
root.title("Adding a label")
root.geometry("500x300")
root.config(bg="light blue")
########### Label ###########################################
label=tk.Label(root, 
               text="Welcome to TKINTER",
               font=("Areail",14))
label.pack(pady=1)  # Alignment in top, middle,bottom , left, right
############### Boolean var ##########
var=tk.BooleanVar()

checkbutton=tk.Checkbutton(root,
                           text="Agree Terms and conditions!",
                          variable=var)
checkbutton.pack(pady=20)
################# Check STatus +++
def check_status():
    if var.get():
        label.config(text="You Agreed to terms and COnditions")
    else:
        label.config(text="you did not agree")

button=tk.Button(root,
                 text="Submit",
                 command=check_status)
        
button.pack(pady=20)
root.mainloop()
```

```python
import tkinter as tk
root=tk.Tk()
root.title("Adding a label")
root.geometry("500x300")
root.config(bg='light blue')
 
l1=tk.Label(root,text='Name:')
l1.grid(row=0,column=0,padx=10,pady=10)
 
l2=tk.Label(root,text='Age:')
l2.grid(row=1,column=0,padx=10,pady=10)
 
e1=tk.Entry(root)
e1.grid(row=0,column=1,padx=10,pady=10)
 
e2=tk.Entry(root)
e2.grid(row=1,column=1,padx=10,pady=10)
 
def show_info():
    name=e1.get()
    age=e2.get()
    print(f"Name:{name},Age:{age}")
    l5=tk.Label(root,text=f"you entered name as {e1.get()} having age as {e2.get()}", font=("Arial",14))
    l5.grid(row=4,column=0,padx=10,pady=10)
 
button=tk.Button(root,text="Submit",command=show_info)
button.grid(row=3,column=0,pady=20)



root.mainloop()
```

```python
# BMI FUnction
def calculate_bmi():
    weight=eval(entry_weight.get())
    height=eval(entry_height.get())
    bmi=weight/(height**2)
    result_label.config(text=f"BMI:{bmi:.2f}")
    if bmi<18:
        category="under weight"
    elif 18<=bmi<24:
        category="Normal weight"
    else:
        category="Over weight"
    category_label.config(text=f"category:{category}")

import tkinter as tk
root=tk.Tk()
root.title("Adding a label")
root.geometry("500x300")
root.config(bg='light blue')
########################## Label ###############################
label=tk.Label(root,
              text="Body Mass Index(BMI) Calculator!",
              font=("Arial",14))
label.pack(pady=10) # adding the label with some pixels
 
weight_label=tk.Label(root,
              text="Weight (Kg)",
              font=("Arial",12))
weight_label.pack(pady=5)
entry_weight=tk.Entry(root)
entry_weight.pack(pady=5)
 
height_label=tk.Label(root,
              text="Height (m)",
              font=("Arial",12))
height_label.pack(pady=5)
entry_height=tk.Entry(root)
entry_height.pack(pady=5)
 
button=tk.Button(root,text="Calulate BMI",command=calculate_bmi)
button.pack(pady=5)
 
result_label=tk.Label(root,text="BMI")
result_label.pack(pady=5)
 
category_label=tk.Label(root,text="Category")
category_label.pack(pady=5)
root.mainloop()
```
